```python
# 01. Ingest Data
# Objective: Download and structure the full text of the KJV Bible.

import pandas as pd
import requests
import io

def load_bible_data():
    print("üì• DOWNLOADING BIBLE TEXT...")
    # URL for a clean JSON/CSV version of KJV
    url = "https://raw.githubusercontent.com/scrollmapper/bible_databases/master/csv/t_kjv.csv"
    
    try:
        content = requests.get(url).content
        df = pd.read_csv(io.StringIO(content.decode('utf-8')))
        
        # Rename cols for clarity: id, b (book), c (chapter), v (verse), t (text)
        df.columns = ['ID', 'Book_ID', 'Chapter', 'Verse', 'Text']
        
        # Create a "Reference" column (e.g., "1:1")
        df['Ref'] = df['Chapter'].astype(str) + ":" + df['Verse'].astype(str)
        
        print(f"‚úÖ Bible Loaded: {len(df)} Verses.")
        return df
    except Exception as e:
        print(f"‚ùå Error: {e}")
        return pd.DataFrame()

df_bible = load_bible_data()

# Save to disk so other notebooks can read it
df_bible.to_csv("bible_master.csv", index=False)
print("üíæ Saved to bible_master.csv")
print(df_bible.head())
```
