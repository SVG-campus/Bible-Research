# 12. The Sin Virus
## Generational Error Propagation

**Objective:** Model the 'Sins of the Fathers' (Exodus 20:5) as recursive code decay.
**Metaphysical Concept:** Genetic/Spiritual Entropy.

```python
import numpy as np
import matplotlib.pyplot as plt

def propagate_sin(generations=10):
    print(f"ðŸ§¬ TRACKING BLOODLINES ({generations} Gens)...")
    
    # Perfect Code (Adam)
    code_integrity = [1.0] 
    
    # The Fall (Gen 1)
    error_rate = 0.05 # Initial corruption
    
    for g in range(generations):
        # Previous generation's integrity
        prev = code_integrity[-1]
        
        # Recursive Decay: Integrity * (1 - error)
        # Error rate compounds (Entropy increases)
        current = prev * (1 - error_rate)
        
        # Variable error (Environment)
        noise = np.random.normal(0, 0.01)
        current += noise
        
        code_integrity.append(current)
        
        # Error rate slightly increases as system degrades
        error_rate *= 1.1
        
    return code_integrity

history = propagate_sin()
final_state = history[-1]

print(f"Original Integrity: 100.0%")
print(f"Final Integrity:    {final_state*100:.2f}%")
print("\nðŸ“‰ CONCLUSION: Without external patching (Grace), the system trends toward 0% integrity (Death).")
```
