```python
# 03. Network Graph
# Objective: Create a simplified interaction graph of major Biblical figures.

import networkx as nx
import matplotlib.pyplot as plt

# Define manual edges for clarity
# (Automated extraction is possible but often messy; manual is cleaner for learning)
relationships = [
    ("Adam", "Eve"), ("Adam", "Abel"), ("Adam", "Cain"), ("Cain", "Abel"),
    ("Noah", "Shem"), ("Noah", "Ham"), ("Noah", "Japheth"),
    ("Abraham", "Sarah"), ("Abraham", "Isaac"), ("Abraham", "Lot"),
    ("Isaac", "Rebekah"), ("Isaac", "Jacob"), ("Isaac", "Esau"),
    ("Jacob", "Rachel"), ("Jacob", "Leah"), ("Jacob", "Joseph"),
    ("Joseph", "Pharaoh"), ("Moses", "Aaron"), ("Moses", "Miriam"),
    ("Moses", "Joshua"), ("David", "Goliath"), ("David", "Jonathan"),
    ("David", "Solomon"), ("David", "Bathsheba"),
    ("Jesus", "Peter"), ("Jesus", "John"), ("Jesus", "James"),
    ("Jesus", "Mary"), ("Jesus", "Judas"), ("Jesus", "Pilate"),
    ("Paul", "Peter"), ("Paul", "Barnabas"), ("Paul", "Timothy")
]

print("üï∏Ô∏è BUILDING GRAPH...")
G = nx.Graph()
G.add_edges_from(relationships)

plt.figure(figsize=(12, 10))
pos = nx.spring_layout(G, seed=42, k=0.5) # k controls node spacing

# Draw Nodes
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color='lightblue', edgecolors='black')

# Draw Edges
nx.draw_networkx_edges(G, pos, width=2, alpha=0.5, edge_color='gray')

# Draw Labels
nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif', font_weight='bold')

plt.title("Key Biblical Relationships Network")
plt.axis('off') # Turn off axis
plt.show()
```
