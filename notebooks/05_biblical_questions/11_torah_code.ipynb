# 11. The Torah Code Matrix
## Equidistant Letter Sequences (ELS)

**Objective:** Determine if 'Encoded' words appear in the Torah at skip sequences more often than in random text.
**The Question:** Is there a hidden watermark in the text?

```python
# 1. SIMULATE TEXT GRIDS
import numpy as np
import random
import string

def search_els(text, target, max_skip=10):
    found = 0
    n = len(text)
    target_len = len(target)
    
    # Brute force search for skip sequences
    for skip in range(1, max_skip + 1):
        for start in range(n - (target_len * skip)):
            # Extract sequence
            seq = text[start : start + (target_len * skip) : skip]
            if seq == target:
                found += 1
    return found

print("üìú SCANNING MANUSCRIPTS...")

# 1. Random Noise (Control Group)
# Generate 100,000 random characters
random_text = ''.join(random.choices(string.ascii_uppercase, k=100000))

# 2. Structured Torah (Simulated)
# We inject a 'Code' artificially to test detection
torah_sim = list(random_text)
# Inject "TORAH" at skip 50 starting at index 100
code = "TORAH"
skip = 50
start = 100
for i, char in enumerate(code):
    torah_sim[start + (i*skip)] = char
torah_text = "".join(torah_sim)

print("‚úÖ Texts Generated.")
```

```python
# 2. RUN SEARCH
target_word = "TORAH"

matches_random = search_els(random_text, target_word, max_skip=100)
matches_torah = search_els(torah_text, target_word, max_skip=100)

print(f"Target: {target_word}")
print(f"Matches in Random Noise: {matches_random}")
print(f"Matches in Torah Sim:    {matches_torah}")

if matches_torah > matches_random:
    print("\nüîç SIGNAL DETECTED: The structured text contains skip sequences not present in noise.")
    print("   Note: While controversial, ELS finding requires P-value testing against 'War and Peace' baselines.")
```
