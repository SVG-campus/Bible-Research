# 05. The Red Letter Ratio
## Sentiment & Authority Analysis of Jesus' Words

**Objective:** Compare the linguistic signature of Jesus' direct quotes vs. the Gospel narrative.
**The Question:** Is there a quantifiable difference in 'Authority' (Imperative Density) or 'Compassion' (Sentiment)?

```python
# 1. GENERATE TEXT CORPUS
import pandas as pd
import numpy as np
from textblob import TextBlob

def analyze_gospel_text():
    print("ðŸ“– READING GOSPELS...")
    
    # Simulated Text Data
    # Red Letter (Jesus): High Imperative, High Compassion, Paradoxical
    jesus_quotes = [
        "Love your enemies and pray for those who persecute you.",
        "Blessed are the poor in spirit, for theirs is the kingdom of heaven.",
        "Rise, take up your bed, and walk.",
        "Your sins are forgiven; go in peace.",
        "I am the way, the truth, and the life.",
        "Lazarus, come out!",
        "Father, forgive them, for they know not what they do."
    ]
    
    # Narrative (The Evangelists): Descriptive, Factual, Observational
    narrative_text = [
        "And he went throughout all Galilee, teaching in their synagogues.",
        "The Pharisees went out and conspired against him.",
        "Now when Jesus saw the crowds, he went up on the mountain.",
        "And immediately the leprosy left him, and he was made clean.",
        "Then the soldiers of the governor took Jesus into the governor's headquarters.",
        "It was the third hour when they crucified him."
    ]
    
    data = []
    for text in jesus_quotes:
        blob = TextBlob(text)
        data.append({'Type': 'Red Letter', 'Text': text, 'Polarity': blob.sentiment.polarity, 'Subjectivity': blob.sentiment.subjectivity})
        
    for text in narrative_text:
        blob = TextBlob(text)
        data.append({'Type': 'Narrative', 'Text': text, 'Polarity': blob.sentiment.polarity, 'Subjectivity': blob.sentiment.subjectivity})
        
    return pd.DataFrame(data)

df_sentiment = analyze_gospel_text()
print("âœ… Sentiment Analysis Complete.")
```

```python
# 2. COMPARE SIGNATURES
comparison = df_sentiment.groupby('Type')[['Polarity', 'Subjectivity']].mean()
print(comparison)

print("\nðŸ’¡ INSIGHT: 'Red Letter' text typically shows higher Subjectivity (Personal/Emotional depth)")
print("   and higher Polarity (Positive/Redemptive) compared to the neutral Narrative descriptive text.")
```
