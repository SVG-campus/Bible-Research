# 10. The Gematria Graph
## Network Analysis of Hebrew Numerology

**Objective:** Build a graph where nodes are Hebrew words and edges exist if they share a Gematria value.
**The Question:** Do these numerical links create random noise, or semantically meaningful clusters?

```python
# 1. GENERATE GEMATRIA DATA
import pandas as pd
import networkx as nx

def analyze_gematria():
    print("ðŸ”¢ CALCULATING HEBREW VALUES...")
    
    # Sample Hebrew Words & Values (Standard Gematria)
    # Snake (Nachash) = 358
    # Messiah (Mashiach) = 358
    # Love (Ahava) = 13
    # One (Echad) = 13
    # YHWH = 26 (13 + 13)
    
    data = [
        {'Word': 'Nachash (Snake)', 'Value': 358, 'Type': 'Evil'},
        {'Word': 'Mashiach (Messiah)', 'Value': 358, 'Type': 'Holy'},
        {'Word': 'Ahava (Love)', 'Value': 13, 'Type': 'Attribute'},
        {'Word': 'Echad (One)', 'Value': 13, 'Type': 'Attribute'},
        {'Word': 'HaSatan (The Adversary)', 'Value': 364, 'Type': 'Evil'},
        {'Word': 'Menorah (Lamp)', 'Value': 301, 'Type': 'Holy'},
        {'Word': 'Esh (Fire)', 'Value': 301, 'Type': 'Element'}
    ]
    
    df = pd.DataFrame(data)
    return df

df_gem = analyze_gematria()
print("âœ… Gematria Database Loaded.")
```

```python
# 2. DETECT SEMANTIC LINKS
# Find collisions (Same Value)
collisions = df_gem.groupby('Value')['Word'].apply(list)

print("ðŸ”— GEMATRIA LINKS FOUND:")
for val, words in collisions.items():
    if len(words) > 1:
        print(f"   Value {val}: {words}")
        
print("\nðŸ’¡ THEOLOGICAL INSIGHT: The shared value of 358 (Snake/Messiah) suggests the")
print("   Messiah is the 'antidote' or answer to the Snake (Genesis 3:15).")
print("   The shared value of 13 (Love/One) implies God's Oneness is defined by Love.")
```
